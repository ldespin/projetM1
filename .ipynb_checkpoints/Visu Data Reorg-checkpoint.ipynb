```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import cv2
from sklearn.metrics import mean_squared_error
from skimage import img_as_float
from skimage.measure import compare_mse as mse
```

```python
f=h5py.File("./data_reorg/val_1/data_1_train.h5","r")
f.keys()
```

```python
def get_imflow(flow, im_shape=(128, 128, 3)):
    hsv = np.zeros(im_shape, dtype=np.uint8)
    hsv[..., 1] = 255
    
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])
    hsv[..., 0] = ang * 180 / np.pi / 2
    hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)

    im = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

    return im
```

```python
data_mouth = f['mouth_3']
flow_mouth = data_mouth[50]

im_mouth = get_imflow(flow_mouth,(64,64,3))

plt.imshow(im_mouth,cmap = 'gray', interpolation = 'bicubic')

plt.show()
```
